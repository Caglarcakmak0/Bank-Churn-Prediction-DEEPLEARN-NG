```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
from sklearn.metrics import classification_report
import numpy as np



# SMOTE ile veri dengesini sağlama
smote = SMOTE(sampling_strategy="minority")
X_sm, y_sm = smote.fit_resample(X, y)

X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.2, random_state=15, stratify=y_sm)

def ANN(X_train, y_train, X_test, y_test, loss):
    model = Sequential([
        Dense(13, input_dim=X_train.shape[1], activation='relu'),
        Dense(10, activation='relu'),
        Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer='adam', loss=loss, metrics=['accuracy'])

    model.fit(X_train, y_train, epochs=100)

    print(model.evaluate(X_test, y_test))

    y_preds = model.predict(X_test)
    y_preds = np.round(y_preds)

    print("Classification Report: \n", classification_report(y_test, y_preds))

    return model, y_preds


```

```python
# Modelin oluşturulması ve tahminlerin alınması
model, y_preds = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy')

# Model ile tahmin yapılması
new_y_preds = model.predict(X_test)
new_y_preds_classes = (new_y_preds > 0.5).astype(int)  # Eşik değeri 0.5 olarak belirlenmiştir

# Sonuçların incelenmesi
print(new_y_preds_classes)
```
